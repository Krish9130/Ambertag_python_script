```python
import pandas as pd

# Display the DataFrame to see the loaded data
print("Original DataFrame:\n", df)

# Step 2: Data Manipulation

# a. Filter rows where Country is 'United States'
us_df = df[df['Country'] == 'United States']
print("\nFiltered DataFrame (Country: United States):\n", us_df)

# b. Group by Gender and calculate the average age
gender_age_avg = df.groupby('Gender')['Age'].mean().reset_index()
print("\nAverage Age by Gender:\n", gender_age_avg)

# c. Sort data by Age in descending order
sorted_df = df.sort_values(by='Age', ascending=False)
print("\nDataFrame sorted by Age (Descending):\n", sorted_df)
```

```python
import matplotlib.pyplot as plt

# Bar plot for average age by gender
plt.figure(figsize=(8, 5))
plt.bar(gender_age_avg['Gender'], gender_age_avg['Age'], color=['skyblue', 'lightgreen'])
plt.title('Average Age by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Age')
plt.show()

# Pie chart for gender distribution
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Gender Distribution')
plt.show()
```
